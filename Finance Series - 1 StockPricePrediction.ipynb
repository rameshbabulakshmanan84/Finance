```python

#Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Tata Global Beverageâ€™s past stock prices.
#Open: Open price of that day
#High: High price of that day
#Low: Low shareprice of that day
#Close : Closing proce of the day
#last : last traded price of the day

```python
#import the data
url="https://raw.githubusercontent.com/mwitiderrick/stockprice/master/NSE-TATAGLOBAL.csv"
#use dataframe
df=pd.read_csv(url)
#print the df
df
```

```python
#top 5 rows
df.head()
```

# STEP 1 : Data preprocessing

```python
#use "open" and "high" as training datasets
train_ds=df.iloc[:,1:2]
```

```python
#print the ds
train_ds
```

```python
#use feature scaler 
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler(feature_range=(0,1))
train_ds_scale=sc.fit_transform(train_ds)
#print the ds
train_ds_scale
```

# Stpe 2: Incorporating Timesteps Into Data


```python
X_train=[]
Y_train=[]
for i in range(60,2035):
    X_train.append(train_ds_scale[i-60:i,0])
    Y_train.append(train_ds_scale[i,0])
X_train=np.array(X_train)
Y_train=np.array(Y_train)

    
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
#Reshape the X_train into 3D
X_train=np.reshape(X_train,(X_train.shape[0],X_train.shape[1],1))
```

```python
print(X_train.shape)
```

# Step 3: create the LSTM Model

```python
from tensorflow.keras.models import Sequential
#Add LSTM layer
from tensorflow.keras.layers import LSTM
#Add Droput to avoid overfitting
from tensorflow.keras.layers import Dropout
#Add dense layer
from tensorflow.keras.layers import Dense
```

# Build the model

```python
model=Sequential()
#add LSTM layer
#units = 50 means 50 dim outputs
#return sequence = True to stack LSTM layers
#input_shape = training dataset shape
model.add(LSTM(units=50,return_sequences=True,input_shape=(X_train.shape[1],1)))
#add a droput layer with 0.2 ->20% of layers dropped
model.add(Dropout(0.2))
#second LSTM layer
model.add(LSTM(units=50,return_sequences=True))
#sedcond dropout
model.add(Dropout(0.2))
#third LSTM
model.add(LSTM(units=50,return_sequences=True))
#third droput
model.add(Dropout(0.2))
#fourth LSTM
model.add(LSTM(units=50))
#fourth droput
model.add(Dropout(0.2))
#add a dense layer
model.add(Dense(units=1))

#compile

```

# Compile the model

```python
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.summary()
```

# Train the model

```python
model.fit(X_train,Y_train,epochs=100,batch_size=32)
```
